```python
# Example data


# Create the CountVectorizer object and fit on the data
count_vectorizer = CountVectorizer()
descriptions = [' '.join(ad['TotalDescp']) for ad in list2]
vectors = count_vectorizer.fit_transform(descriptions)

# Define a new token array
new_tokens = ['this', 'is', 'the', 'first',"maison" ,'ad']

# Transform the new token array into a bag-of-words vector
new_description = ' '.join(new_tokens)
new_vector = count_vectorizer.transform([new_description])

# Calculate the cosine distances between the new vector and the data vectors
distances = cosine_distances(new_vector, vectors)

# Print the distances
print(distances)

```

```python
print(type(distances))
lst = distances.tolist()
flat_list = [elem for sublist in lst for elem in sublist]

print(flat_list) 
```

```python
df['distances']=flat_list
df_sorted = df.sort_values('distances', ascending=True)
```
